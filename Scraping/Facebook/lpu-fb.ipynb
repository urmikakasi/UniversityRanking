```python
import time 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import nltk 
import io 
import unicodedata 
import numpy as np 
import re 
import csv
import string 
from numpy import linalg 
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
from nltk.tokenize import sent_tokenize, word_tokenize 
from nltk.tokenize import PunktSentenceTokenizer 
from nltk.tokenize import PunktSentenceTokenizer 
from nltk.corpus import webtext 
from nltk.stem.porter import PorterStemmer 
from nltk.stem.wordnet import WordNetLemmatizer 
```

```python
df=pd.read_csv('res1.csv')

lpu=df[733:1419]
lpu=lpu[['timestamp','name','message','review']]
lpu
```

```python
vals=[]
for index,row in lpu.iterrows():
    m=row['message']
    r=row['review']
    if isinstance(m,float) and isinstance(r,float):
        vals.append(np.NaN)
    else:
        vals.append(0)
lpu['del']=vals
lpu.dropna(subset=['del'],inplace=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
vals=[]
#print(iitb)
df=lpu[['message','review']]
#print(df)
for index,row in df.iterrows():
    v=row["message"]
    #print(row)
    if isinstance(v,float):
        y=row["review"]
        #print(y)
        x=re.search("(recommends)",str(y))
        #print(x)
        z=re.search("(reviewed)",str(y))
        
        if x is not None:
            vals.append(0.05)
        
        elif z is not None:
            print(z.string)
            num=re.search('[5|4|3|2|1]',z.string)
            print(num.group())
            if num.group()=='5':
                vals.append(0.3)
            elif num.group()=='4':
                vals.append(0.1)
            elif num.group()=='3':
                vals.append(0)
            elif num.group()=='2':
                vals.append(-0.1)
            elif num.group()=='1':
                vals.append(-0.3)
            #vals.append(0.05)
        if x is None and z is None:
            vals.append(-0.05)
            
        #iitb['compound']=0
        #vals.append(0)
        
    else:
        
        val=analyzer.polarity_scores(v)['compound']
        vals.append(val)
#print(vals)
lpu['compound'] = vals 
```

```python
lpu=lpu.drop('del',axis=1)
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

lpu['Sentiment'] = lpu['compound'].apply(lambda x: sentiment(x))
lpu
```

```python
lpu['pred'] = pd.cut(lpu['compound'], bins=5, labels=[1, 2, 3, 4, 5])
vals=[]
for index, row in lpu.iterrows():
    if (row['compound']==0.05):
        row['pred']=4
        vals.append(4)
        
    elif row['compound']==-0.05:
        row['pred']=2
        vals.append(2)
    elif row['Sentiment']=='Positive' and row['pred']<=3:
        vals.append(4)
    elif row['Sentiment']=='Negative' and row['pred']>=3:
        vals.append(2)
    elif row['Sentiment']=='Neutral' and row['pred']!=3:
        vals.append(3)
    else:
        vals.append(row['pred'])
lpu['pred']=vals    
```
