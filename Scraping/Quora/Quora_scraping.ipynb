```python
import urllib
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
import time
```

```python
pesu_df1= pd.read_csv('C:\\Users\\KIRHIKAGURUMURTHY\\Desktop\\PES stuff\\Projects\\Uni_ranking\\quora-scraper\\quora-data.csv')
```

```python
pesu_df = pd.DataFrame({'question': [],'answers':[]})
url = pesu_df1['Question link']
pesu_df1 =pesu_df1.rename(columns={"Title": "question"})
```

```python
for i in url:
    try:
        page = requests.get(i)
        soup = BeautifulSoup(page.content, "html.parser")
        question = soup.find('span', 
            {'class':      'ui_qtext_rendered_qtext'})
        answers = soup.find_all('div', 
            attrs={'class':    'ui_qtext_expanded'})
        for answer in answers:
            pesu_df = pesu_df.append({'question': question.text,
                'answers': answer.text
                }, ignore_index=True)
    except requests.exceptions.ConnectionError:
        page.status_code = "Connection refused"
```
